```python


# Define the paths
image_dir = 'images/Images'
annotation_dir = 'annotations/Annotation'

# Initialize an empty DataFrame to store the data
# data = {'ID': [],
#         'Breed': [],
#         'Image': [],
#         'Path': [],
#         'Width': [],
#         'Height': [],
#         'Xmin': [],
#         'Xmax': [],
#         'Ymin': [],
#         'Ymax': []}
# df = pd.DataFrame(data)
raw_data = []
```

```python
#print(os.listdir(image_dir))
#print(os.listdir(annotation_dir))
```

```python
# Traverse through image directory

i = 0
for breed_folder in os.listdir(image_dir):
#     if i > 10:  # Used for testing a subset
#         break
    breed_path = os.path.join(image_dir, breed_folder)
    if os.path.isdir(breed_path):
        breed = breed_folder.split('-')[-1].lower()  # Lowercase for uniformity
        print("Breed is:", breed) # FIXME: remove or comment
        
        # Identify each image in the subfolder
        for image_file in os.listdir(breed_path):
            if image_file.endswith('.jpg'):
                image_id = os.path.splitext(image_file)[0]
                image_path = os.path.join(breed_path, image_file)
                
                # Find corresponding annotation file
                annotation_file = os.path.join(annotation_dir, breed_folder, image_id)
                # print("Annotation file is:", annotation_file) # FIXME: remove or comment
                # Check if annotation file exists
                if os.path.exists(annotation_file):
                    # Read annotation file
                    with open(annotation_file, 'r') as f:
                        lines = f.readlines()
                    
                    # Extract annotation information
                    width = int(int(lines[7].split('>')[1].split('<')[0]))
                    #print(lines[7], width) # FIXME: remove
                    height = int(lines[8].split('>')[1].split('<')[0])
                    #print(lines[8], height) # FIXME: remove
                    xmin = int(lines[18].split('>')[1].split('<')[0])
                    #print(lines[18], xmin) # FIXME: remove
                    ymin = int(lines[19].split('>')[1].split('<')[0])
                    xmax = int(lines[20].split('>')[1].split('<')[0])
                    ymax = int(lines[21].split('>')[1].split('<')[0])
                    
                    # Open image and crop
                    img = Image.open(image_path)
                    img_cropped = img.crop((xmin, ymin, xmax, ymax))
                    #img.show()
                    #img_cropped.show()
                    
                    # Add the new entry to the list
                    raw_data.append({'ID': image_id,
                                      'Breed': breed,
                                      'Image': img_cropped,
                                      'Path': image_path,
                                      'Width': width,
                                      'Height': height,
                                      'Xmin': xmin,
                                      'Xmax': xmax,
                                      'Ymin': ymin,
                                      'Ymax': ymax})
                    
#     i+= 1
                    
```

```python
df = pd.DataFrame(raw_data)
df.head()
display(df)
```

# Model

```python
# Todo: Takes forever to run, not good

import numpy as np
import pandas as pd
from PIL import Image
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications import NASNetLarge
from vit_keras import vit

# Function to extract features using NASNet
def extract_features_nasnet(image):
    nasnet_model = NASNetLarge(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    img = image.resize((224, 224))
    img_array = np.expand_dims(np.array(img) / 255.0, axis=0)
    features = nasnet_model.predict(img_array)
    return features.flatten()

# Function to extract features using ViT
def extract_features_vit(image):
    vit_model = vit.vit_b32(image_size=256, activation=None, pretrained=True, include_top=False)
    img = image.resize((256, 256))
    img_array = np.expand_dims(np.array(img) / 255.0, axis=0)
    features = vit_model.predict(img_array)
    return features.flatten()

# Load DataFrame with "Image" column containing PIL images
# Assuming df is your DataFrame
# df = pd.read_csv('your_data.csv')

# Extract features using NASNet and ViT
nasnet_features = df['Image'].apply(extract_features_nasnet)
vit_features = df['Image'].apply(extract_features_vit)

# Concatenate features
all_features = np.concatenate([nasnet_features.values, vit_features.values], axis=1)

# Perform PCA for dimensionality reduction
pca = PCA(n_components=100)  # Adjust number of components as needed
reduced_features = pca.fit_transform(all_features)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(reduced_features, df['Label'], test_size=0.2, random_state=42)

# Train SVM classifier
svm_classifier = SVC()
svm_classifier.fit(X_train, y_train)

# Evaluate SVM classifier
accuracy = svm_classifier.score(X_test, y_test)
print("Accuracy:", accuracy)
```
